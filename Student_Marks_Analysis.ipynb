```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Marks': [80, 70, 90, 85, 75]
}

# Create DataFrame
df = pd.DataFrame(data)

# Show data
print("Dataset:")
print(df)

# Basic analysis
print("\nStatistics:")
print(df.describe())

# Plot graph
df.plot(x='Name', y='Marks', kind='bar')

plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
